# A review of numerical simulations for mono-strut-based $H_2$ injection into supersonic airflows 

A total of 14 papers available on google scholar were referred to for conducting a comprehensive review of all the numerical simulations conducted so far for strut-based hydrogen-injected supersonic-combustion. Each paper's major takeaways has been summarized. Click on the headings to be redirected to the paper.

 # [1. Gerlinger and Bruggeman, Stuttgart](https://sci-hub.do/https://arc.aiaa.org/doi/pdf/10.2514/2.5559)

**Type of simulation**: 2-D DNS considering 3-species *non-reacting* flow ($H_2, O_2, N_2$) + $q-\omega$ turbulence model \
**Inflow conditions**: 2.05 bar, 240K air + 1.95 bar, 260K $H_2$ both @ Mach 1 \
**Geometry type**: Normal (rectangular) with a diamond-strut \

**Mesh resolution**: 0.5 micron \
**Solver used**: In-house FVM code based on the Euler conservation laws \
**Results**: Emphasis on the mixing characteristics and the influence of the strut geometry

# [2. Aravind and Rajiv Kumar, BIT Mesra](https://www.sciencedirect.com/science/article/abs/pii/S0360319919301661)

**Type of simulation**: RANS with $k-\epsilon$ for turbulence-closure. $\beta-PDF$ for **non-premixed** combustion with a flamelet-based TCI (Turbulence-Chemistry Interaction) \
**Solver Used**: ANSYS Fluent\
**Inflow conditions**: See below:

**Geometry type**: DLR Combustor\

**Mesh resolution**: 0.5 micron \
**Results**: Emphasis on the mixing characteristics and the influence of the strut geometry\

# [4. Kumaran and Babu, IIT Madras](https://www.sciencedirect.com/science/article/abs/pii/S0010218009000108)

**Key-takeaways**: Discusses the importance of multi-step detailed chemistry on the numerical simulations for supersonic combution. \
**Simulation type**: 3-D compressible reacting RANS using Spallart-Almaras turbulence model \
**Solver used**: ANSYS Fluent\
**Geometry, Inflow conditions**: Staged combustor with an expanding geometry:

**Chemistry & Kinetics**: 37-reaction + 9-species $H_2$-combustion\
**Mesh resolution**: 0.7M-1M cells with 55 avg wall $y+$

# [5.Wu et.al, Chinese Academy of Sciences](https://www.tandfonline.com/doi/full/10.1080/00102202.2017.1365847?needAccess=true)

**Type of simulation**: 2-D reacting LES\
**Geometry & Inflow**: DLR Combustor\

**Mesh size**: 0.7M cells (2-D) and 3M in 3-D\
**Chemistry model**: 9-species 19 reactions\
**Solver**: AstroFOAM (OpenFOAM)\
**Results**: combustion-stages, sensitivity analysis of the chemistry-model used

# [6.Wu et.al., Chinese Academy of Science, flame stabilization](https://www.sciencedirect.com/science/article/abs/pii/S1540748918305157)

**Summary**: Reference to the previous paper with other operating conditions to investigate modes of flame stabilization\
**Key takeaways**:Identified 2 flame-stabilization modes: an attached flame mode and a lifted flame mode\
**Operating conditions**: DLR combustor geometry with $\phi$ ranging from 0.022 to 0.11, Temperature inflow ranging from 607K to 2141K
